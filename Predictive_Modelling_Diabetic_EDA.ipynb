```python
df.loc[df['PeopleHH'] > 3 , 'large_family'] = 1
df.loc[df['PeopleHH'] <= 3 , 'large_family'] = 0
df.loc[~df['EduLevelAdult'].isin(['College/AA', 'College/Above', '>HighSchool']) , 'college_educated'] = 0
df.loc[df['EduLevelAdult'].isin(['College/AA', 'College/Above', '>HighSchool']) , 'college_educated'] = 1
df.loc[df['EduLevelAdult'].isin(['Unknown', 'DontKnow']) , 'college_educated'] = 2
df.loc[df['FamilyIncome'].isin(['<5k','5-10k', '10-15k','15-20k', '20-25k', '25-35k','35-45k', '45-55k']) , 'FamilyIncome'] = 'Low'
df.loc[df['FamilyIncome'].isin(['55-65k','65-75k', '75-99k']) , 'FamilyIncome'] = 'Medium'
df.loc[df['FamilyIncome'].isin(['>100k']) , 'FamilyIncome'] = 'High'
```

```python
df.loc[df['USStayLength'].isin(['Unknown','DontKnow', 'Refused']) , 'USStayLength'] = 'Unknown'
df.loc[df['USStayLength'].isin(['<1', '1-5', '5-10']) , 'USStayLength'] = 'Short Term (< 10 Years)'
df.loc[df['USStayLength'].isin(['10-15', '15-20','20-30', '30-40','40-50', '>50']) , 'USStayLength'] = 'Long Term (> 10 Years)'
```

```python
df.loc[df['Citizenship'] != 'USCitizen', 'Citizenship'] = 0
df.loc[df['Citizenship'] == 'USCitizen', 'Citizenship'] = 1
```

```python
changed_col = ['PeopleHH', 'EduLevelAdult']
df.drop(columns=changed_col, inplace = True)
```

```python
df.loc[df['MaritalStatus'].isin(['Unknown','DontKnow', 'Refused']) , 'MaritalStatus'] = 'Unknown'
df.loc[df['MaritalStatus'].isin(['Divorced', 'Separated', 'Widowed']) , 'MaritalStatus'] = 'Separated/Divorced'
df.loc[df['MaritalStatus'].isin(['Married', 'LivingIn']) , 'MaritalStatus'] = 'Married'
df.loc[df['MaritalStatus'].isin(['NeverMarried']) , 'MaritalStatus'] = 'Single'
```

```python
df['high_risk'] = 0
df.loc[(df['AgeYears'] <= 20) & (df['Body Mass Index (kg/m**2)'] >= 25) & (df['Glycohemoglobin (%)'] >= 5.7) & (df['Two Hour Glucose(OGTT) (mg/dL)'] >= 140), 'high_risk'] = 1
df.loc[(df['AgeYears'] > 20) & (df['AgeYears'] <= 40) & (df['Body Mass Index (kg/m**2)'] >= 25) & (df['Glycohemoglobin (%)'] >= 5.7) & (df['Two Hour Glucose(OGTT) (mg/dL)'] >= 120), 'high_risk'] = 1
df.loc[(df['AgeYears'] > 40) & (df['Body Mass Index (kg/m**2)'] >= 25) & (df['Glycohemoglobin (%)'] >= 5.7) & (df['Two Hour Glucose(OGTT) (mg/dL)'] >= 100), 'high_risk'] = 1
```
